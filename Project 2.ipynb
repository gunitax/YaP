Скорректируем положительные и отрицательные значения в колонке days_employed

```python
# функция для коррекции days_employed
def correct_days_employed(el):
    if el > 0:
        return el / 24
    elif el < 0:
        return -el
    else:
        return el

data['days_employed'] = data['days_employed'].apply(correct_days_employed)
```

```python
# процент ошибочных записей в поле income_type
err_income = data[data['total_income'].isna()]['children'].count()
print('Процент ошибочных записей в поле total_income {:.4%}'.format(err_income / rows_in_dataset))
```

```python
data['total_income'] = data['total_income'].fillna(0)
# заполним пропуски дохода средним значением в разрезе типа дохода,
# если среднее отличается от медианы в разрезе типа дохода более чем на 10%, иначе медианой
income_types = list(data[data['total_income'] == 0]['income_type'].unique())
for income_type in income_types:
    subdata = data[(data['income_type'] == income_type) & (data['total_income'] != 0)]['total_income']
    mean = subdata.mean()
    median = subdata.median()
    try:
        if mean / median > 1.1:
            res = median
        else:
            res = mean
    except: # медиана = 0
        res = mean                                                    
    data.loc[(data['income_type'] == income_type) & (data['total_income'] == 0), 'total_income'] = res
```

```python
# На основании диапазонов, указанных ниже, создайте столбец total_income_category с категориями:

#     0–30000 — 'E';
#     30001–50000 — 'D';
#     50001–200000 — 'C';
#     200001–1000000 — 'B';
#     1000001 и выше — 'A'.

# функция возвращает значение категории в зависимости от переданного дохода 

def category(income):
    if income <= 30000:
        return 'E'
    elif income <= 50000:
        return 'D'
    elif income < 200000:
        return 'C'   
    elif income < 1000000:
        return 'B'   
    else:
        return 'A'
    
data['total_income_category'] = data['total_income'].apply(category)
```

```python
# количество значений по категориям дохода
data['total_income_category'].value_counts()
```

```python
# Создайте функцию, которая на основании данных из столбца purpose сформирует новый столбец
# purpose_category, в который войдут следующие категории: 
# 'операции с автомобилем',
# 'операции с недвижимостью',
# 'проведение свадьбы',
# 'получение образования'

# определим словарь, в котором будут в зависимости от категории подстроки, появление которых будет относить
# purpose к той или иной категории, заполнение сделано вручную на основании уникальных значений столбца purpose
purpose_dict = {
    'операции с автомобилем': ['автомобил'],
    'операции с недвижимостью': ['жиль', 'недвиж'],
    'проведение свадьбы': ['свадьб'],
    'получение образования' : ['образова']
}

def get_purpose_category(purpose):
    for key, value in purpose_dict.items():
        for substr in value:
            if purpose.find(substr) > -1:
                return key
    return None

data['purpose_category'] = data['purpose'].apply(get_purpose_category)
data['purpose_category'].value_counts()
```
