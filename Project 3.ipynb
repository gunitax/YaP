–ø—Ä–µ–¥–ø–æ–ª–æ–∂—É —á—Ç–æ –ø–ª–æ—â–∞–¥—å –∫—É—Ö–Ω–∏ –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –ø—Ä–æ–ø–æ—Ä—Ü–∏–æ–Ω–∞–ª—å–Ω–∞ –æ–±—â–µ–π –ø–ª–æ—â–∞–¥–∏ –∫–≤–∞—Ä—Ç–∏—Ä—ã, –¥–ª—è –≤—ã–≤–µ–¥–µ–Ω–∏—è –∫–æ—ç—Ñ—Ñ–∏—Ü–µ–Ω—Ç–∞ –ø–æ—Å—á–∏—Ç–∞–µ–º  –º–µ–¥–∏–∞–Ω—É –ø–æ –∑–∞–ø–æ–ª–Ω–µ–Ω–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏—è–º –∏ –∑–∞–ø–æ–ª–Ω–∏–º –ø–ª–æ—â–∞–¥—å –∫—É—Ö–Ω–∏ 

```python
kitchen_koef_median = df['kitchen_area'].median() / df['total_area'].median()
print(f'–°–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –º–µ–¥–∏–∞–Ω –ø–æ –ø–ª–æ—â–∞–¥–∏ –∫—É—Ö–Ω–∏: {kitchen_koef_median}')

def calc_kitchen(row):    
    if row['kitchen_area'] != row['kitchen_area']: # —Ç—Ä—é–∫ –ø–∏—Ç–æ–Ω–∞ –¥–ª—è NaN
        row['kitchen_area'] = row['total_area'] * kitchen_koef_median
        if row['kitchen_area'] < 5:
            row['kitchen_area'] = 5
    return row

df = df.apply(calc_kitchen, axis=1)
```

```python
deleted_words = ['–ø–æ—Å–µ–ª–æ–∫', '–ø–æ—Å—ë–ª–æ–∫', '—Å–µ–ª–æ', '–≥–æ—Ä–æ–¥—Å–∫–æ–π', '–¥–µ—Ä–µ–≤–Ω—è', '–∫–æ—Ç—Ç–µ–¥–∂–Ω—ã–π', \
             '–≥–æ—Ä–æ–¥—Å–∫–æ–≥–æ —Ç–∏–ø–∞', '—Å—Ç–∞–Ω—Ü–∏–∏', '–ø—Ä–∏ –∂–µ–ª–µ–∑–Ω–æ–¥–æ—Ä–æ–∂–Ω–æ–π', '—Å–∞–¥–æ–≤–æ–¥—á–µ—Å–∫–æ–µ  —Ç–æ–≤–∞—Ä–∏—â–µ—Å—Ç–≤–æ', \
            '–Ω–µ–∫–æ–º–º–µ—Ä—á–µ—Å–∫–æ–µ', '—Å–∞–¥–æ–≤–æ–µ —Ç–æ–≤–∞—Ä–∏—â–µ—Å—Ç–≤–æ']

def correct_name(name):
    new_name = name
    for deleted_word in deleted_words:
        new_name = new_name.replace(deleted_word,'').strip()
    return new_name


df['locality_name'] = df['locality_name'].apply(correct_name)
```

```python
# –¥–ª—è —Å–µ—Ä–∏–∏ –ø–æ–ª—É—á–∞–µ–º –ø–ª—é—Å –º–∏–Ω—É—Å 1,5 –º–µ–∂–∫–≤–∞–Ω—Ç–∏–ª—å–Ω—ã—Ö —Ä–∞–∑–º–∞—Ö–∞ + –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –±–∞—Ä–æ–≤
def get_hist_data(serie):
    desc = serie.describe()
    
    value_min = desc[3]
    value_max = desc[7]
    
    q1 = desc[4]
    q3 = desc[6]
    
    body = q3 - q1
    left = max(value_min, q1 - 1.5 * body)
    right = min(value_max, q3 + 1.5 * body)
    
    bins = right - left
    
    if bins > 100:
        bins = 100
    elif bins <=1:
        bins = bins * 10 + 1

    bins = int(bins)
        
    return (left, right, bins)

```
